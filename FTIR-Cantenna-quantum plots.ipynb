```python
import numpy as np
import matplotlib.pyplot as plt 
#loading data

tirapproach = np.loadtxt("ftir-cantenna-quantum-data-2.txt",usecols=[1],unpack=True)
tirapart = np.loadtxt("ftir-cantenna-quantum-data-2.txt",usecols=[2],unpack=True)
ftirapproach = np.loadtxt("ftir-cantenna-quantum-data-2.txt",usecols=[3],unpack=True)
ftirapart = np.loadtxt("ftir-cantenna-quantum-data-2.txt",usecols=[4],unpack=True)

tirapproach2 = np.loadtxt("ftir-cantenna-quantum-data-2.txt",usecols=[5],unpack=True)
tirapart2 = np.loadtxt("ftir-cantenna-quantum-data-2.txt",usecols=[6],
                       unpack=True)
ftirapproach2 = np.loadtxt("ftir-cantenna-quantum-data-2.txt",usecols=[7],unpack=True)
ftirapart2 = np.loadtxt("ftir-cantenna-quantum-data-2.txt",usecols=[8],unpack=True)

#loading the data


dist=[0,2,4,6,8,10,12]
```

```python
#plotting reflected signal strength in dbm

plt.subplot(1,2,1)
plt.plot(dist,tirapproach,'o',label= "Decreasing distance ", color = '#D81B60', markersize=25, marker ="v")
plt.plot(dist,tirapart,'o',label= "Increasing distance ", color = '#1E88E5', markersize=25, marker ="d")
plt.plot(dist,tirapproach2,'o',label= "Decreasing distance  [second attempt]", color = '#FFC107', markersize=25)
plt.plot(dist,tirapart2,'o',label= "Increasing distance  [second attempt]", color = '#1EF1CD', markersize=25, marker ="v")

plt.xlabel("Separation in Centimeters",fontsize = 30)
plt.ylabel("Signal strength in dBm",fontsize = 30)
plt.yticks(fontsize=25)
plt.xticks(fontsize=25)
plt.legend(fontsize = 25)
plt.ticklabel_format(style='plain')
#plotting reflected signal strength in milli-Watts

plt.subplot(1,2,2)
plt.plot(dist,10**(tirapproach/10)*1000000,'o',label= "Decreasing distance ", color = '#D81B60', markersize=25, marker ="v")
plt.plot(dist,10**(tirapart/10)*1000000,'o',label= "Increasing distance ", color = '#1E88E5', markersize=25, marker ="d")
plt.plot(dist,10**(tirapproach2/10)*1000000,'o',label= "Decreasing distance  [second attempt]", color = '#FFC107', markersize=25)
plt.plot(dist,10**(tirapart2/10)*1000000,'o',label= "Increasing distance  [second attempt]", color = '#1EF1CD', markersize=25, marker ="v")

plt.xlabel("Separation in Centimeters",fontsize = 30)
plt.ylabel("Signal strength in milli-Watts",fontsize = 30)
plt.rcParams['figure.figsize'] = (35, 12)
plt.yticks(fontsize=25)
plt.xticks(fontsize=25)
plt.legend(fontsize = 25)
plt.ticklabel_format(style='plain')

plt.suptitle('Reflection side of the prism',fontweight ="bold", fontsize = 35)



plt.savefig("reflectionW.pdf",bbox_inches='tight')
```

```python
phi=45
#angle of incidence
n=1.45
#refractive index
lambdaa= 12.5
#wavelength
f=[]
for d in [np.linspace(0, 12, num=1000)]:
    y=(((2*3.1415*d)/12.5)*(((n**2)*(np.sin(45))**2)-1)**(0.5))
    f.append(y) 



alphaper= ((((n**2) - 1)/(2*n))**2) * (1/(((np.cos(phi))**2) * ((n**2) * (np.sin(phi))**2 -1)))

T=(1/(alphaper*(((np.sinh(f))**2)+1)))

#print(f)
#print(T)

#theoretical values for transmission coefficient 
```

```python
#T.shape=(1)








d=np.linspace(0, 12, num=1000)


T=T.reshape(-1)

og=10**((-55)/10)*1000000
#dBm to milliwatt conversion of direct measurement 


#plotting transmission coeff, theoretical and observered 
print(og)
plt.plot(dist,(10**((ftirapproach)/10)*1000000)/og,'o', label= "Decreasing distance", color = '#D81B60', markersize=25,  marker ="v")
plt.plot(dist,(10**((ftirapart)/10)*1000000)/og,'o',label= "Increasing distance", color = '#1E88E5',  markersize=25,  marker ="d")
plt.plot(dist,(10**((ftirapproach2)/10)*1000000)/og,'o', label= "Decreasing distance [second attempt]", color = '#FFC107', markersize=25)
plt.plot(dist,(10**((ftirapart2)/10)*1000000)/og,'o',label= "Increasing distance  [second attempt]", color = '#1EF1CD', markersize=25, marker ="^")
plt.plot(d,T,label= "Theoretical Values", color = 'black')
plt.xlabel("Separation in Centimeters",fontsize = 30)
plt.ylabel("Transmission Coefficient",fontsize = 30)
plt.rcParams['figure.figsize'] = (24, 18)
plt.yticks(fontsize=25)
plt.xticks(fontsize=25)
plt.legend(fontsize = 25)

plt.suptitle('Transmission Coefficients With respect to Distance between prisms',fontweight ="bold", fontsize = 35)

#plt.tight_layout()
plt.savefig("transmissionWcoef.pdf",bbox_inches='tight')





```
